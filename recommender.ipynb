# Text pre-processing

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer

def _removeNonAscii(s):
    return "".join(i for i in s if  ord(i)<128)

def make_lower_case(text):
    return text.lower()

def remove_stop_words(text):
    text = text.split()
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops]
    text = " ".join(text)
    return text

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

def remove_punctuation(text):
    tokenizer = RegexpTokenizer(r'\w+')
    text = tokenizer.tokenize(text)
    text = " ".join(text)
    return text

# call the functions and clean the text for the title, short_description, and description columns
data['title'] = data['title'].apply(_removeNonAscii)
data['title'] = data['title'].apply(make_lower_case)
data['title'] = data['title'].apply(remove_stop_words)
data['title'] = data['title'].apply(remove_punctuation)
data['title'] = data['title'].apply(remove_html)

data['short_description'] = data['short_description'].apply(_removeNonAscii)
data['short_description'] = data['short_description'].apply(make_lower_case)
data['short_description'] = data['short_description'].apply(remove_stop_words)
data['short_description'] = data['short_description'].apply(remove_punctuation)
data['short_description'] = data['short_description'].apply(remove_html)

data['description'] = data['description'].apply(_removeNonAscii)
data['description'] = data['description'].apply(make_lower_case)
data['description'] = data['description'].apply(remove_stop_words)
data['description'] = data['description'].apply(remove_punctuation)
data['description'] = data['description'].apply(remove_html)

data
```

```python
corpus_title = data['title'].tolist()
corpus_short_description = data['short_description'].tolist()
corpus_description = data['description'].tolist()

def get_corpus(text):
    words = []
    for i in text:
        text = [word for word in i.split()]
        words.append(text)
    return words

corpus_title = get_corpus(corpus_title)
corpus_short_description = get_corpus(corpus_short_description)
corpus_description = get_corpus(corpus_description)
```

# Search based on the keywords

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import numpy as np

def get_average_word2vec(tokens_list, vector=glove_model, generate_missing=False, k=300):
    if len(tokens_list) < 1:
        return np.zeros(k)
    
    vectorized = []
    for word in tokens_list:
        if word in vector:
            vectorized.append(vector[word])
        elif generate_missing:
            vectorized.append(np.random.rand(k))
    
    if not vectorized:
        return np.zeros(k)
    
    # Only compute the mean if there are non-zero vectors
    if len(vectorized) > 0:
        return np.mean(vectorized, axis=0)
    else:
        return np.zeros(k)



def get_word2vec_embeddings(vectors, corpus, generate_missing=False):
    embeddings = []
    for title in corpus:
        averaged_embedding = get_average_word2vec(title, vectors, generate_missing=generate_missing)
        embeddings.append(averaged_embedding)
    return embeddings


from sklearn.metrics.pairwise import cosine_similarity

def recommend_news_based_on_keyword(keyword, vectors, data):
    keyword = keyword.split()
    keyword = get_average_word2vec(keyword, vectors)
    
    # Calculate vectors for title
    # cut down the title into lists of words before applying the get_average_word2vec functio
    data['title_vector'] = data['title'].apply(lambda x: get_average_word2vec(x.split(), vectors))
    data['short_desc_vector'] = data['short_description'].apply(lambda x: get_average_word2vec(x.split(), vectors))
    data['desc_vector'] = data['description'].apply(lambda x: get_average_word2vec(x.split(), vectors))

    # Compute cosine similarity for each title vector
    data['title_similarity'] = data.apply(lambda row: cosine_similarity([row['title_vector']], [keyword])[0][0], axis=1)
    data['short_desc_similarity'] = data.apply(lambda row: cosine_similarity([row['short_desc_vector']], [keyword])[0][0], axis=1)
    data['desc_similarity'] = data.apply(lambda row: cosine_similarity([row['desc_vector']], [keyword])[0][0], axis=1)

    # Combine similarities with weights
    title_weight = 0.7
    short_desc_weight = 0.2
    desc_weight = 0.1

    data['combined_similarity'] = (title_weight * data['title_similarity'] +
                                      short_desc_weight * data['short_desc_similarity'] +
                                      desc_weight * data['desc_similarity'])

    # Sort by title similarity
    data = data.sort_values(by='title_similarity', ascending=False)
    
    return data

recommend_news_based_on_keyword('ukraine', glove_model, data)[:10]
```

# Search based on author's name

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_news_based_on_author(author, vectors, data, recent=False):
    author = author.split()
    author = get_average_word2vec(author, vectors)
    
    # Calculate vectors for title
    # cut down the title into lists of words before applying the get_average_word2vec functio
    data['author_vector'] = data['author'].apply(lambda x: get_average_word2vec(x.split(), vectors))

    # Compute cosine similarity for each title vector
    data['author_similarity'] = data.apply(lambda row: cosine_similarity([row['author_vector']], [author])[0][0], axis=1)

    # Sort by title similarity
    data = data.sort_values(by='author_similarity', ascending=False)

    # Sort by date_created if recent is True
    if recent:
        data = data.sort_values(by=['author_similarity', 'date_created'], ascending=[False, False])

    
    return data

recommend_news_based_on_author('matthew monks', glove_model, data, recent=True)[:10]
```

```python
def recommend_news_based_on_keywords_and_preferences(keywords, vectors, data, user_preferences=None):
    # Initialize combined similarity column
    data['combined_similarity'] = 0.0
    
    # Iterate through each keyword and its score
    for keyword, score in keywords.items():
        # Calculate average word2vec representation for the keyword
        keyword_vec = get_average_word2vec(keyword.split(), vectors)
        
        # Calculate cosine similarity for each feature with the keyword
        data['title_similarity'] = data.apply(lambda row: cosine_similarity([row['title_vector']], [keyword_vec])[0][0], axis=1)
        data['short_desc_similarity'] = data.apply(lambda row: cosine_similarity([row['short_desc_vector']], [keyword_vec])[0][0], axis=1)
        data['desc_similarity'] = data.apply(lambda row: cosine_similarity([row['desc_vector']], [keyword_vec])[0][0], axis=1)

        # Combine similarities with weights
        title_weight = 0.7
        short_desc_weight = 0.2
        desc_weight = 0.1

        data['combined_similarity'] += (title_weight * data['title_similarity'] +
                                        short_desc_weight * data['short_desc_similarity'] +
                                        desc_weight * data['desc_similarity']) * score
    
    # Sort by combined similarity
    data = data.sort_values(by='combined_similarity', ascending=False)

    data = data.sort_values(by='date_created', ascending=False)

    
    # Prioritize articles based on user preferences for categories
    if user_preferences:
        category_weights = user_preferences
        category_max_score = max(category_weights.values())
        for category, weight in category_weights.items():
            data.loc[data['category'] == category, 'combined_similarity'] *= (weight / category_max_score)
    
    # Sort again by adjusted combined similarity
    data = data.sort_values(by='combined_similarity', ascending=False)
    
    return data

recommend_news_based_on_keywords_and_preferences(user_keywords, glove_model, data, user_categories)[:20]
```

# Basic functions

```python
def preprocessing(text):
    text = _removeNonAscii(text)
    text = make_lower_case(text)
    text = remove_stop_words(text)
    text = remove_punctuation(text)
    text = remove_html(text)
    text = text.split()
    return text

def update_like_keywords(text, user_keywords, current_time, weight=0.1):
    text = preprocessing(text)
    
    for keyword in text:
        if keyword in user_keywords:
            user_keywords[keyword]['score'] += user_keywords[keyword]['score'] * weight
            user_keywords[keyword]['last_modified'] = current_time
            if user_keywords[keyword]['score'] > 5:
                user_keywords[keyword]['score'] = 5
        else:
            user_keywords[keyword] = {'score': weight, 'last_modified': current_time}
    return user_keywords

def update_like_categories(category, user_categories, current_time, weight=0.1):
    if category in user_categories:
        user_categories[category]['score'] += user_categories[category]['score'] * weight
        user_categories[category]['last_modified'] = current_time
        if user_categories[category]['score'] > 5:
            user_categories[category]['score'] = 5
    else:
        user_categories[category] = {'score': weight, 'last_modified': current_time}
    return user_categories


def update_dislike_keywords(text, user_keywords, current_time, weight=0.1):
    text = preprocessing(text)
    
    for keyword in text:
        if keyword in user_keywords:
            user_keywords[keyword]['score'] -= user_keywords[keyword]['score'] * weight
            user_keywords[keyword]['last_modified'] = current_time
            if user_keywords[keyword]['score'] < 0:
                user_keywords[keyword]['score'] = -0
        else:
            user_keywords[keyword] = {'score': -weight, 'last_modified': current_time}
    return user_keywords

def update_dislike_categories(category, user_categories, current_time, weight=0.1):
    if category in user_categories:
        user_categories[category]['score'] -= user_categories[category]['score'] * weight
        user_categories[category]['last_modified'] = current_time
        if user_categories[category]['score'] < 0:
            user_categories[category]['score'] = 0
    else:
        user_categories[category] = {'score': -weight, 'last_modified': current_time}
    return user_categories


def delete_overflow_keywords(user_keywords):
    if len(user_keywords) > 1000:
        sorted_keywords = sorted(user_keywords.items(), key=lambda x: x[1]['last_modified'], reverse=False)
        for keyword, data in sorted_keywords[:100]:
            del user_keywords[keyword]
    return user_keywords

def delete_overflow_categories(user_categories):
    if len(user_categories) > 1000:
        sorted_categories = sorted(user_categories.items(), key=lambda x: x[1]['last_modified'], reverse=False)
        for category, data in sorted_categories[:100]:
            del user_categories[category]
    return user_categories
```

# Viewing history

```python
import json
from collections import OrderedDict
from datetime import datetime

user = json.load(open('data/user.json'))

def update_view_history(user_data, article_history):
    user_keywords = user_data['user_keywords']
    user_categories = user_data['user_categories']

    current_time = datetime.now().timestamp()

    for article in article_history:
        category = article['category']

        # Update keywords
        update_like_keywords(article['title'], user_keywords, current_time, weight=0.1)

        # Update categories
        update_like_categories(category, user_categories, current_time, weight=0.1)
            
    sorted_keywords = sorted(user_keywords.items(), key=lambda x: x[1]['last_modified'], reverse=True)
    sorted_categories = sorted(user_categories.items(), key=lambda x: x[1]['last_modified'], reverse=True)

    updated_keywords = OrderedDict()
    updated_categories = OrderedDict()

    for keyword, score in sorted_keywords:
        updated_keywords[keyword] = score

    for category, score in sorted_categories:
        updated_categories[category] = score

    updated_keywords = delete_overflow_keywords(updated_keywords)
    updated_categories = delete_overflow_categories(updated_categories)

    user_data['user_keywords'] = updated_keywords
    user_data['user_categories'] = updated_categories
    
    with open('data/user.json', 'w') as f:
        json.dump(user_data, f, indent=4)

update_view_history(user, user['user_history']['articles'])
```

```python
def set_feedback(user_data, article, feedback):
    current_time = datetime.now().timestamp()

    # Update keywords
    if feedback.lower() == 'like':
        if article not in user_data['liked_articles']:
            user_data['user_keywords'] = update_like_keywords(article['title'], user_data['user_keywords'], current_time, weight=0.8)
    elif feedback.lower() == 'dislike':
        if article not in user_data['disliked_articles']:
            user_data['user_keywords'] = update_dislike_keywords(article['title'], user_data['user_keywords'], current_time, weight=0.8)

    # Update category
    if feedback.lower() == 'like':
        if article not in user_data['liked_articles']:
            user_data['user_categories'] = update_like_categories(article['category'], user_data['user_categories'], current_time, weight=0.8)
    elif feedback.lower() == 'dislike':
        if article not in user_data['disliked_articles']:
            user_data['user_categories'] = update_dislike_categories(article['category'], user_data['user_categories'], current_time, weight=0.8)

    # Check if liked_articles/disliked_articles keys exist
    if 'liked_articles' not in user_data:
        user_data['liked_articles'] = []
    if 'disliked_articles' not in user_data:
        user_data['disliked_articles'] = []

    # Memorize the liked/disliked article
    if feedback.lower() == 'like' and article not in user_data['liked_articles']:
        user_data['liked_articles'].append(article)
    elif feedback.lower() == 'dislike' and article not in user_data['disliked_articles']:
        user_data['disliked_articles'].append(article)
    else:
        print("Invalid feedback. Please provide 'like' or 'dislike'.")

    # Sort user_keywords and user_categories by last modified time
    user_data['user_keywords'] = OrderedDict(sorted(user_data['user_keywords'].items(), key=lambda x: x[1]['last_modified'], reverse=True))
    user_data['user_categories'] = OrderedDict(sorted(user_data['user_categories'].items(), key=lambda x: x[1]['last_modified'], reverse=True))

    with open('data/user.json', 'w') as f:
        json.dump(user_data, f, indent=4)
    

user = json.load(open('data/user.json'))
article = articles[np.random.randint(0, len(articles))]
print(article)
set_feedback(user, article, 'like')
```
