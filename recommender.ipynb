```python
def check_missing_values(df):
    # Check for NaN, missing, or NaT values
    missing_rows = df[df.isnull().any(axis=1)]

    if not missing_rows.empty:
        print("Deleting rows with missing values...")
        df.dropna(inplace=True)
        print("Rows with missing values have been deleted.")
    else:
        print("There are no missing values in the DataFrame.")

# Example usage:
check_missing_values(data)
data
```

# Text pre-processing

```python
import re
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer

def _removeNonAscii(s):
    return "".join(i for i in s if  ord(i)<128)

def make_lower_case(text):
    return text.lower()

def remove_stop_words(text):
    text = text.split()
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops]
    text = " ".join(text)
    return text

def remove_html(text):
    html_pattern = re.compile('<.*?>')
    return html_pattern.sub(r'', text)

def remove_punctuation(text):
    tokenizer = RegexpTokenizer(r'\w+')
    text = tokenizer.tokenize(text)
    text = " ".join(text)
    return text

# call the functions and clean the text for the title, short_description, and description columns
data['title'] = data['title'].apply(_removeNonAscii)
data['title'] = data['title'].apply(make_lower_case)
data['title'] = data['title'].apply(remove_stop_words)
data['title'] = data['title'].apply(remove_punctuation)
data['title'] = data['title'].apply(remove_html)

data['short_description'] = data['short_description'].apply(_removeNonAscii)
data['short_description'] = data['short_description'].apply(make_lower_case)
data['short_description'] = data['short_description'].apply(remove_stop_words)
data['short_description'] = data['short_description'].apply(remove_punctuation)
data['short_description'] = data['short_description'].apply(remove_html)

data['description'] = data['description'].apply(_removeNonAscii)
data['description'] = data['description'].apply(make_lower_case)
data['description'] = data['description'].apply(remove_stop_words)
data['description'] = data['description'].apply(remove_punctuation)
data['description'] = data['description'].apply(remove_html)

data
```

```python
corpus_title = data['title'].tolist()
corpus_short_description = data['short_description'].tolist()
corpus_description = data['description'].tolist()

def get_corpus(text):
    words = []
    for i in text:
        text = [word for word in i.split()]
        words.append(text)
    return words

corpus_title = get_corpus(corpus_title)
corpus_short_description = get_corpus(corpus_short_description)
corpus_description = get_corpus(corpus_description)
```

```python
from gensim.models import KeyedVectors

glove_model = KeyedVectors.load_word2vec_format('gloves/converted_vectors300.txt', binary=False)
```

# Search based on the keywords

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string
import numpy as np

def get_average_word2vec(tokens_list, vector=glove_model, generate_missing=False, k=300):
    if len(tokens_list) < 1:
        return np.zeros(k)
    
    vectorized = []
    for word in tokens_list:
        if word in vector:
            vectorized.append(vector[word])
        elif generate_missing:
            vectorized.append(np.random.rand(k))
    
    if not vectorized:
        return np.zeros(k)
    
    # Only compute the mean if there are non-zero vectors
    if len(vectorized) > 0:
        return np.mean(vectorized, axis=0)
    else:
        return np.zeros(k)



def get_word2vec_embeddings(vectors, corpus, generate_missing=False):
    embeddings = []
    for title in corpus:
        averaged_embedding = get_average_word2vec(title, vectors, generate_missing=generate_missing)
        embeddings.append(averaged_embedding)
    return embeddings


from sklearn.metrics.pairwise import cosine_similarity

def recommend_news_based_on_keyword(keyword, vectors, data):
    keyword = keyword.split()
    keyword = get_average_word2vec(keyword, vectors)
    
    # Calculate vectors for title
    # cut down the title into lists of words before applying the get_average_word2vec functio
    data['title_vector'] = data['title'].apply(lambda x: get_average_word2vec(x.split(), vectors))
    data['short_desc_vector'] = data['short_description'].apply(lambda x: get_average_word2vec(x.split(), vectors))
    data['desc_vector'] = data['description'].apply(lambda x: get_average_word2vec(x.split(), vectors))

    # Compute cosine similarity for each title vector
    data['title_similarity'] = data.apply(lambda row: cosine_similarity([row['title_vector']], [keyword])[0][0], axis=1)
    data['short_desc_similarity'] = data.apply(lambda row: cosine_similarity([row['short_desc_vector']], [keyword])[0][0], axis=1)
    data['desc_similarity'] = data.apply(lambda row: cosine_similarity([row['desc_vector']], [keyword])[0][0], axis=1)

    # Combine similarities with weights
    title_weight = 0.7
    short_desc_weight = 0.2
    desc_weight = 0.1

    data['combined_similarity'] = (title_weight * data['title_similarity'] +
                                      short_desc_weight * data['short_desc_similarity'] +
                                      desc_weight * data['desc_similarity'])

    # Sort by title similarity
    data = data.sort_values(by='title_similarity', ascending=False)
    
    return data

recommend_news_based_on_keyword('ukraine', glove_model, data)[:10]
```

# Search based on author's name

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_news_based_on_author(author, vectors, data, recent=False):
    author = author.split()
    author = get_average_word2vec(author, vectors)
    
    # Calculate vectors for title
    # cut down the title into lists of words before applying the get_average_word2vec functio
    data['author_vector'] = data['author'].apply(lambda x: get_average_word2vec(x.split(), vectors))

    # Compute cosine similarity for each title vector
    data['author_similarity'] = data.apply(lambda row: cosine_similarity([row['author_vector']], [author])[0][0], axis=1)

    # Sort by title similarity
    data = data.sort_values(by='author_similarity', ascending=False)

    # Sort by date_created if recent is True
    if recent:
        data = data.sort_values(by=['author_similarity', 'date_created'], ascending=[False, False])

    
    return data

recommend_news_based_on_author('matthew monks', glove_model, data, recent=True)[:10]
```

# Keywords and Categories plus Recent

```python
def recommend_news_based_on_keywords_and_preferences(keywords, vectors, data, user_preferences=None):
    # Initialize combined similarity column
    data['combined_similarity'] = 0.0
    
    # Iterate through each keyword and its score
    for keyword, score in keywords.items():
        # Calculate average word2vec representation for the keyword
        keyword_vec = get_average_word2vec(keyword.split(), vectors)
        
        # Calculate cosine similarity for each feature with the keyword
        data['title_similarity'] = data.apply(lambda row: cosine_similarity([row['title_vector']], [keyword_vec])[0][0], axis=1)
        data['short_desc_similarity'] = data.apply(lambda row: cosine_similarity([row['short_desc_vector']], [keyword_vec])[0][0], axis=1)
        data['desc_similarity'] = data.apply(lambda row: cosine_similarity([row['desc_vector']], [keyword_vec])[0][0], axis=1)

        # Combine similarities with weights
        title_weight = 0.7
        short_desc_weight = 0.2
        desc_weight = 0.1

        data['combined_similarity'] += (title_weight * data['title_similarity'] +
                                        short_desc_weight * data['short_desc_similarity'] +
                                        desc_weight * data['desc_similarity']) * score
    
    # Sort by combined similarity
    data = data.sort_values(by='combined_similarity', ascending=False)

    data = data.sort_values(by='date_created', ascending=False)

    
    # Prioritize articles based on user preferences for categories
    if user_preferences:
        category_weights = user_preferences
        category_max_score = max(category_weights.values())
        for category, weight in category_weights.items():
            data.loc[data['category'] == category, 'combined_similarity'] *= (weight / category_max_score)
    
    # Sort again by adjusted combined similarity
    data = data.sort_values(by='combined_similarity', ascending=False)
    
    return data

recommend_news_based_on_keywords_and_preferences(keywords, glove_model, data, user_preferences)[:20]
```
